About the Dataset:  

    The dataset contains 120,000 records of mortality data for older individuals, sourced from the World Health Organization (WHO) website. It provides information on the leading causes of death among older populations across different countries and regions over various years. This dataset aims to analyze mortality trends and demographic factors related to older individuals' health.  
    
For accessing the dataset, refer: https://platform.who.int/data/maternal-newborn-child-adolescent-ageing/indicator-explorer-new/MCA/mortality-ranking-in-older-people---top-10-causes-(country)


Features:  

    Indicator: Describes the specific health metric being reported. 

    Year: The year the data was collected or reported. 

    Country: The country where the data was collected. 

    WHO region: The World Health Organization region classification for the country.

    World bank income group: The income classification of the country by the World Bank. 

    Age group: The age range of the population being reported on.

    Cause: The cause of death being reported.

    Measure: The type of measure being reported, i.e., "Rank". 

    Sex: The gender of the population being reported.

    Value Numeric: The numeric value representing the measure. 

    Value String: The string representation of the measure. 

Aim of the Project:  

    This project aims to do detailed analysis of causes of death across different demographics and regions.

# 1. What are the top 10 causes of death in older people across different years and regions?

```python
# identify the top 10 causes of death for older adults
top_causes = who_df1['cause'].value_counts().nlargest(10).index

# filter the dataset to include only the top 10 causes of death
top_causes_df = who_df1[who_df1['cause'].isin(top_causes)]

# create a count plot for these top 10 causes across different years and regions
plt.figure(figsize=(14, 6))
sns.set_style("whitegrid")
sns.countplot(data=top_causes_df, x='cause', hue='region', palette='viridis', order=top_causes)
plt.xlabel('Count')
plt.xticks(rotation=90)
plt.ylabel('Cause of Death')
plt.grid(True)
plt.title('Top 10 Leading Causes of Death for Older Adults Across Different Regions')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Observation:

    The top 10 causes of death are:

    Cardiovascular diseases
    Ischemic heart disease
    Stroke
    Malignant neoplasms (cancer)
    Respiratory diseases
    Lower respiratory infections
    Digestive diseases
    Haemorrhagic stroke
    Ischemic stroke
    
    The prevalence of CVD, IHD and Stroke is the highest and have similar trends in Europe, Africa and America. 
    CVD, IHD are the leading causes of death in Eastern Mediterranian region
    CVD, IHD, Stroke and Malignant Neoplasms are the leading causes of death in Western pacific and South-East Asia region 

```python
# find WHO region with max cardiovascular diseases related deaths

cv_deaths = who_df1[who_df1['cause'] == 'Cardiovascular diseases']

country_cv_deaths= cv_deaths.groupby('region')['deaths'].sum().reset_index()

highest_deaths_country = country_cv_deaths.loc[country_cv_deaths['deaths'].idxmax()]

print(f"WHO region with the highest number of deaths from cardiovascular diseases among older people is {highest_deaths_country['region']} with {highest_deaths_country['deaths']} deaths.")
```

```python
# Find the country in Europe with the highest number of deaths from cardiovascular diseases

europe_cv_deaths = cv_deaths[cv_deaths['region'] == 'Europe']

country_cv_deaths = europe_cv_deaths.groupby('country')['deaths'].sum().reset_index()

highest_cv_deaths_country = country_cv_deaths.loc[country_cv_deaths['deaths'].idxmax()]

print(f"Country in Europe with the highest number of deaths from cardiovascular diseases among older people is {highest_cv_deaths_country['country']} with {highest_cv_deaths_country['deaths']} deaths.")
```

```python
# find country with max cardiovascular diseases related deaths

cv_deaths = who_df1[who_df1['cause'] == 'Cardiovascular diseases']

country_cv_deaths= cv_deaths.groupby('country')['deaths'].sum().reset_index()

highest_deaths_country = country_cv_deaths.loc[country_cv_deaths['deaths'].idxmax()]

print(f"Country with the highest number of deaths from cardiovascular diseases among older people is {highest_deaths_country['country']} with {highest_deaths_country['deaths']} deaths.")
```

# 2. What is the trend of average number of deaths over the years?

```python
# calculate average deaths per year
average_deaths_per_year = who_df1.groupby('year')['deaths'].mean().reset_index()

# plot the trend using lineplot
plt.figure(figsize=(12, 4))
sns.set_style("whitegrid")
sns.lineplot(data=average_deaths_per_year, x='year', y='deaths', marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Average Deaths')
plt.title('Trend of Average Deaths per Year')

plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.grid(True)
plt.show()
```

```python
# plot the top causes of death in 2007

deaths_2007 = who_df1[who_df1['year'] == 2007]
total_deaths_2007 = deaths_2007['deaths'].sum()

deaths_by_cause_2007 = deaths_2007.groupby('cause')['deaths'].sum()
percent_deaths_2007 = (deaths_by_cause_2007 / total_deaths_2007) * 100

percent_deaths_df = pd.DataFrame(percent_deaths_2007).reset_index()
percent_deaths_df.columns = ['cause', '%deaths']
percent_deaths_df = percent_deaths_df.sort_values(by='%deaths', ascending=False)
print('Top 5 causes of death in 2007')
display(percent_deaths_df.head(5))

plt.figure(figsize=(15, 3))
sns.set_style("whitegrid")
sns.barplot(x='cause',y='%deaths',data=percent_deaths_df)
plt.xticks(rotation=90)
plt.grid(True)
```

Observation:

    2007 witnessed the highest average number of deaths over the years from 2000-2019 with Haemorrhagic stroke and Respiratory diseases being the top causes accounting for 8.08% and 7.95% deaths respectively
    Lowest deaths were reported in 2015
    

# 3. What is the trend of cause of deaths over the years?

```python
# group year and cause from the filtered df -'top_causes_df'
trend_df = top_causes_df.groupby(['year', 'cause']).size().reset_index(name='count')

# plot the trend using a line plot
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")
sns.lineplot(data=trend_df, x='year', y='count', hue='cause', marker='o')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Trend of Top 10 Causes of Death for Older Adults Over the Years')
plt.legend(title='Cause of Death', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.show()
```

Observation:
  
    Cardiovascular diseases, Ischemic heart disease and Stroke seems to have decreased slightly from 2011
    Ischemic stroke has been reduced slightly over the years
    Malignant neoplasms have slightly increased over the years
    Haemorrhagic stroke and respiratory diseases have significantly reduced from 2000 - 2019
    Digestive diseases have increased slightly during 2005-2009 but the trend showed reduction in deaths due to digestive diseases after 2010
    Respiratory infections and Lower Respiratory infections have shown a similar trend over the years and have shown a slight decrease during 2012
    All the causes have shown a decrease in the deaths during the year 2012

# 4. What is the most common cause of death for each age group?

```python
# identify the most common cause of death for each age group

common_causes = who_df1.groupby(['age_group', 'cause']).size().reset_index(name='count')

most_common_causes = common_causes.loc[common_causes.groupby('age_group')['count'].idxmax()]

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
bars = sns.barplot(data=most_common_causes, x='age_group', y='count', hue='cause', dodge=False, palette='viridis')

plt.xlabel('Age Group')
plt.ylabel('Number of Deaths')
plt.title('Most Common Cause of Death for Each Age Group Among Older People')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.legend(title='Cause of Death', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()

print(f"Percentage of deaths in age group 60-70: {len(who_df1[who_df1['age_group']=='60-70'])/len(who_df1['age_group'])*100:.2f} %")
```

Observation:

    Number of deaths is the highest in age category of 60-70
    Reason for deaths in all age categories of old people are Cardiovascular diseases

# 5. Is there any difference between cause of death in both genders?

```python
# identify the top 10 causes of death for males and females

gender_deaths = who_df1.groupby(['sex', 'cause'])['deaths'].sum().reset_index()


top_10_causes_male = gender_deaths[gender_deaths['sex'] == 'Male'].nlargest(10, 'deaths')
top_10_causes_female = gender_deaths[gender_deaths['sex'] == 'Female'].nlargest(10, 'deaths')

print("Top 10 causes of death in older males:")
display(top_10_causes_male[['cause', 'deaths']])
print("\nTop 10 causes of death in older females:")
display(top_10_causes_female[['cause', 'deaths']])

```

```python
# plotting the causes of death in older males & older females

merged_top_10=pd.concat([top_10_causes_male,top_10_causes_female],axis=0)

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.barplot(data=merged_top_10, x='cause', y='deaths', hue='sex', palette='viridis')
plt.xlabel('Cause of Death')
plt.ylabel('Number of Deaths')
plt.title('Number of Deaths by Cause and Sex')
plt.xticks(rotation=90)
plt.legend(title='Sex')
plt.tight_layout()
plt.show()
```

Observation:

    Respiratory diseases are the major cause of death in older males followed by Stroke
    Haemorrhagic stroke is the major cause of death in older females followed by Ischaemic stroke
    Deaths due to COPD was reported predominantly in older males while deaths due to Diabetes mellitus was reported solely in older females

```python
print(f"Percentage of deaths in males due to Respiratory diseases : {len(top_10_causes_male[top_10_causes_male['cause']=='Respiratory diseases'])/len(top_10_causes_male)*100:.2f} %")
print(f"Percentage of deaths in females due to Haemorrhagic stroke : {len(top_10_causes_female[top_10_causes_female['cause']=='Haemorrhagic stroke'])/len(top_10_causes_female)*100:.2f} %")
```

# 6. Does income status has any effect on death cause?

```python
# visualize income distribution across regions

income_distribution = who_df1.groupby(['region', 'income_group']).size().reset_index(name='count')

income_distribution_grouped = income_distribution.pivot(index='region', columns='income_group', values='count')

plt.figure(figsize=(12, 5))
sns.set_style("whitegrid")
income_distribution_grouped.plot(kind='bar', stacked=True, cmap='plasma')
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Income Distribution Across Regions')
plt.xticks(rotation=90)
plt.legend(title='Income Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Observation:

    High income groups are prevalent in America followed by Europe and Western Pacific region
    Europe has the highest number of people exclusively with higher and upper middle income groups
    South-East Asia has exclusively lower middle income groups
    Africa has lower middle income and low income as the dominant income groups

```python
# visualize the causes of death by income status

income_cause = top_causes_df.groupby(['income_group', 'cause'])['deaths'].sum().reset_index()

plt.figure(figsize=(12, 4))
sns.set_style("whitegrid")
sns.barplot(data=income_cause, x='cause', y='deaths', hue='income_group', palette='Set2')
plt.xticks(rotation=90)
plt.xlabel('Cause of Death')
plt.ylabel('Number of Deaths')
plt.title('Causes of Death by Income Group')
plt.legend(title='Income Group')
plt.show()
```
